```python
import preprocessor as p
import pandas as pd
import numpy as np
import re

data_dir = "../../data/"
tweets_f = "tweets.csv"

output_f = "tweets_cleaned.csv"

tweets_df = pd.read_csv(data_dir + tweets_f)
tweets_df['preprocessed_text'] = tweets_df.apply(lambda row: p.clean(row.text) if isinstance(row.text, str) else None, axis = 1)
tweets_df.head()
```

```python
tweets_df.to_csv(data_dir + output_f)
```
