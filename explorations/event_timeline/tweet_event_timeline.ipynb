```python
tweets_df = pd.read_csv('../input/russian-troll-tweets/tweets.csv')
tweets_df.head()

import datetime as dt
import pandas as pd
election_important_dates = pd.read_csv("/kaggle/input/2016electiondates/2016election_important_dates.csv")
election_important_dates.head()
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

tweets_df['created_datetime'] = pd.to_datetime(tweets_df['created_str'])
num_tweets_by_date = tweets_df['created_datetime'].groupby(tweets_df['created_datetime'].dt.floor('d')).size().reset_index(name='count')
num_tweets_by_date.head()
```

```python
figure(num=None, figsize=(20, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(num_tweets_by_date['created_datetime'], num_tweets_by_date['count'])
for index, row in election_important_dates.iterrows():
    plt.axvline(pd.to_datetime(row['Date']), color='gray')
```
